```python
# Helper functions for basic arithmetic operations
def add(a, b):
  return a + b

def subtract(a, b):
  return a - b

def multiply(a, b):
  return a * b

def divide(a, b):
  if b == 0:
    return "Error: Cannot divide by zero!"
  return a / b

def calculator():
  print("david biscochos calculator")

  # Get first number with input validation
  num1 = None
  while num1 is None:
    try:
      num1_str = input("enter number: ")
      num1 = float(num1_str) # Allow floating-point numbers
    except ValueError:
      print("Invalid input. Please enter a valid number.")

  # Get second number with input validation
  num2 = None
  while num2 is None:
    try:
      num2_str = input("enter another number: ")
      num2 = float(num2_str) # Allow floating-point numbers
    except ValueError:
      print("Invalid input. Please enter a valid number.")

  # Get operation with input validation
  operation = None
  while operation not in ['+', '-', '*', '/']:
    operation = input("Choose operation (+, -, *, /): ").strip()
    if operation not in ['+', '-', '*', '/']:
      print("Invalid operation. Please choose from +, -, *, /.")

  result = None
  if operation == '+':
    result = add(num1, num2)
  elif operation == '-':
    result = subtract(num1, num2)
  elif operation == '*':
    result = multiply(num1, num2)
  elif operation == '/':
    result = divide(num1, num2)

  print(result)

# Start the calculator
calculator()
```

```python
num1 = None
while num1 is None:
    try:
        num1 = float(input("Enter 1st number: "))
    except ValueError:
        print("Please put real number")
num2 = None
while num2 is None:
    try:
        num2 = float(input("Enter 2nd number: "))
    except ValueError:
        print("Please put real number")
num3 = None
while num3 is None:
    try:
        num3 = float(input("Enter 3rd number: "))
    except ValueError:
        print("Please put real number")

def find_maxnum(numbers):
    max_num = numbers[0]

    for num in numbers:
        if num > max_num:
            max_num = num

    return max_num

numbers = [num1, num2, num3]
print(find_maxnum(numbers))
```
